## Matplotlib PyPlot Syntax (plt)

```python
%matplotlib inline
import matplotlib.pyplot as plt

sample_data = [1,4,2,5,3,6]

plt.plot(sample_data)
plt.title('Simple Graph')
plt.xlabel('x axis label')
plt.ylabel('y axis label')
plt.show()

more_data = [6,3,5,2,4,1]
plt.plot(more_data)
plt.title('Simple Graph')
plt.xlabel('other label')
plt.ylabel('some other label')
plt.show()
```

## Matplotlib's fig, ax Syntax (also known as object oriented syntax)

```python
fig, ax = plt.subplots()

fig.patch.set(facecolor = 'green')

ax.plot(sample_data)
ax.set(facecolor='yellow')
ax.set_title('Graph Title')
ax.set_ylabel('my y label')
ax.set_xlabel('my x label')
```

## Reproduce with real data

```python
df = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/inconvenient-sequel/ratings.csv')

df.head()
```

```python
df.tail()
```

```python
# Convert timestamps strings to datetime objects
df['timestamp'] = pd.to_datetime(df['timestamp'])
```

```python
# Use timestamp as the unique index identifier
# to select rows by timestamp
df.set_index('timestamp', inplace=True)
```

```python
# Grab the rows corresponding to the last day
lastday = df['2017-08-29']
```

```python
# Get the demographic breakdowns for all IMDb users on the last day
lastday_filtered = lastday[lastday['category'] =='IMDb users']
lastday_filtered.tail()
```

```python
#Grab the very last line (latest timestamp) of IMDb user ratings
final = lastday_filtered.tail(1)
final.T
```

```python
# Grab percentage columns
pct_columns = ['1_pct', '2_pct', '3_pct', '4_pct', '5_pct',
               '6_pct', '7_pct', '8_pct', '9_pct', '10_pct']
final[pct_columns].T
```

```python
# Rest the index to be numeric and rename the percent column 
plot_data = final[pct_columns].T
plot_data.index = range(1,11)
plot_data.columns = ['percent']
plot_data
```

```python
fig, ax = plt.subplots()
fig.patch.set(facecolor='white')

ax.bar(x=range(1,11), height=plot_data['percent'], width=0.9, color='#ED713A')
ax.set(facecolor='white')

# Set the title and subtitle
ax.set_title("'An Inconvenient Sequel: Truth To Power' is divisive", 
             fontsize = 12, 
             x=.35, 
             y=1.1,
             fontweight='bold')
ax.text(s='IMDb ratings for the film as of Aug. 29', fontsize=12, x=-1.85, y=42)

# Set the x and y axes labels
ax.set_xlabel('Rating', fontsize=9)
ax.set_ylabel('Percent of total votes', fontsize=9)

# Fix the x and y axis tick marks and grid
ax.set_xticklabels(range(1,11), rotation=0)
ax.set_xticks(range(1,11))
ax.set_yticklabels(range(0,50,10))
ax.set_yticks(range(0,50,10))
plt.box(on=None)

# Add % sign to y ticker
import matplotlib.ticker as ticker
ax.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=100, decimals=0))

ax.grid(color='#E6E6E6');
```

```python
#Load data
daily_guest = pd.read_csv('https://raw.githubusercontent.com/fivethirtyeight/data/master/daily-show-guests/daily_show_guests.csv')

print(daily_guest.shape)
daily_guest.head()
```

```python
daily_guest['Group'].unique().tolist()
```

```python
# Create feature with categories
def categories(string):
  acm = ['Acting','Comedy','Musician']
  media = ['Media', 'media']
  gandp = ['Politician','Political Aide','Government','Military']
  if string in acm:
    return 'Acting, Comedy & Music'
  if string in media:
    return 'Media'
  if string in gandp:
    return 'Government and Politics'
  else:
    return 'Other'

daily_guest['Category'] = daily_guest['Group'].apply(categories)

daily_guest.head()
```

```python
table = pd.crosstab(daily_guest['YEAR'], 
                    daily_guest['Category'], 
                    normalize=False)
table
```

```python
# Create Total column
table['total']=table.sum(axis=1)
table.reset_index(inplace=True)

table
```

```python
# Calculate percentage of appearance for category for the features we are graphing
final = table
final['Acting, Comedy & Music'] = table['Acting, Comedy & Music']/final['total']
final['Government and Politics'] = table['Government and Politics']/final['total']
final['Media'] = table['Media']/final['total']
final
```

```python
fig, ax = plt.subplots()

ax.plot(final['YEAR'], final['Acting, Comedy & Music'], color='#1A8ABB',linewidth=2)
ax.plot(final['YEAR'], final['Government and Politics'], color='#ED713A',linewidth=2)
ax.plot(final['YEAR'], final['Media'], color='#7F2A7E',linewidth=2)

ax.set_title("Who Got To Be On 'The Daily Show'?", 
             fontsize = 12, 
             x=0.2, 
             y=1.1,
             fontweight='bold')
ax.text(s='Occupation of guest, by year', fontsize=12, x=1996, y=1.05)
ax.text(s='Acting, Comedy & Music', 
        fontsize=12, 
        color='#1A8ABB', 
        x=2001, y=.8,
        fontweight='bold')
ax.text(s='Media', fontsize=12, color='#7F2A7E', x=2007, y=.5, fontweight='bold')
ax.text(s='Government and Politics', 
        fontsize=12, 
        color='#ED713A', 
        x=2008.5, 
        y=.09, 
        fontweight='bold')

# Fix the x and y axis tick marks and grid
plt.xticks(range(2000,2016,4))
ax.set_xticklabels(['2000',"'04","'08","'12"], rotation=0)
ax.set_yticks([0,.25,.5,.75,1])

ax.yaxis.set_major_formatter(ticker.PercentFormatter(xmax=1, decimals=0));
```
